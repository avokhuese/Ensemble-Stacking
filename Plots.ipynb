```python
data_paths=['No_Augmentation.csv',
 'Jitter_Augmentation.csv',
 'MagW_Augmentation.csv',
 'Permutation_Augmentation.csv',
 'Scale_Augmentation.csv',
 'Rotation_Augmentation.csv',
 'WinW_Augmentation.csv']
titles=[]
for i in data_paths:
  titles.append( i.split('.')[0] )
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
for title, data_path in zip(titles, data_paths):
  data_path=os.path.join('/content/drive/MyDrive/New TS DA', data_path)
  df=pd.read_csv(data_path)
  df.replace('Humidty', 'Humidity', inplace=True)
  df=df[['Data',	'Model','MSE']]
  pd.pivot_table(df, values='MSE', index=['Data'], columns=['Model']).plot(kind='bar', figsize=(12,6))
  plt.ylabel('MSE')
  title='MSE for LSTM, RNN and Stacked Models Across all Dataset (' +  title +    ')'
  plt.title(title)
```
